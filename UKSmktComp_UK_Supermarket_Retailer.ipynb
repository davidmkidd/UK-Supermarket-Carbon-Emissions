## Purpose of this notebook

Before working with carbon emissions data, it is important to understand the scale and diversity of UK supermarkets. Retailers differ substantially in size, business model, and mode of operation. These differences, in turn, influence energy use, carbon emissions, and how emissions data should be interpreted.

## Retailer Size

Larger businesses typically consume more energy and generate higher absolute emissions as a result of their scale. At the same time, larger operations may benefit from economies of scale that improve efficiency.

Business 'size' may be measured using several metrics including,

* Total Revenue
* Number of Stores
* Store Area

Two questions should be asked on these data,

1. How do retailers vary along each dimension?
2. How do the dimensions covary?

These questions will be answered using plots, correlation statistics, and linear regression.

### Comparing Size Metrics

The three size metrics show both consistencies and differences across retailers. Tesco is the largest and Ocado the smallest according to all three measures, while Asda has a comparatively small number of stores and the Co‑op a much larger estate.

Plotting revenue agaist the size variables reveal bivariate relationships between these variables.

Ocado is excluded from this analysis as their revenue is independent of store.

```r
# dpylr filter() selects records where retailer_code is NOT EQUAL (!=) to "OCADO"
data_comparison <- data %>%
  filter(retailer_code != "OCADO")

# In aes(), colour = retailer_code sets the column that defines the colour to display symbols
# geom_point() plots points
# scale_colour_manual() maps retailer_code to the hex code for that category

ggplot(data_comparison, aes(x = store_2023, y = revenue_2023, colour = retailer_code)) +
  geom_point() +
  ggtitle("Retailer Revenue and Store Number with Regression Line") +
  xlab("Store Number") +
  ylab("Revenue million £") +
  scale_colour_manual(name = "Retailer", values = pal) +
  theme_classic()
```

There may be a weak postive correlation between revenue and store number, so let's see if the values correlate.

```r
# Calculate Spearman rank correlation (retained original request)
cor.test(data$revenue_2023, data$store_2023, method = 'spearman')
```

Yes they correlate, but is the relationship linear?

```r
# Linear model of revenue by store number
model <- lm(revenue_2023 ~ store_2023, data = data_comparison)

# geom_smooth() adds a linear regression line and standard error intervals to a plot
ggplot(data_comparison, aes(x = store_2023, y = revenue_2023, color = retailer_code)) +
  geom_point() +
  geom_smooth(method = "lm", se = TRUE, color = "black") + # Add linear regression line
  ggtitle("Retailer Revenue and Store Number with Regression Line") +
  xlab("Store Number") +
  ylab("Revenue million £") +
  scale_colour_manual(name = "Retailer", values = pal) +
  theme_classic()
```

```r
# Get the linear regression model fit statistics
summary(model)
```

Revenue is significantly positively linear with store number, but the model explains just 42.7% of variation in revenue with a low significance of p = 0.040.

Now let's look at revenue and area.

```r
ggplot(data_comparison, aes(x = area_2023, y = revenue_2023, color = retailer_code)) +
  geom_point() +
  ggtitle("Retailer Revenue and Store Number with Regression Line") +
  xlab("Store Area 1000 m2") +
  ylab("Revenue million £") +
  scale_colour_manual(name = "Retailer", values = pal) +
  theme_classic()
```

The relationship appears to be storngly linear.

```r
# Calculate Spearman rank correlation (retained original request)
cor.test(data$revenue_2023, data$area_2023, method = 'spearman')
```

Spearman rank confirms revenue strongly correlates with store area.

```r
# Linear model
model <- lm(revenue_2023 ~ area_2023, data = data_comparison)

ggplot(data_comparison, aes(x = area_2023, y = revenue_2023, color = retailer_code)) +
  geom_point() +
  geom_smooth(method = "lm", se = TRUE, color = "black") + # Add linear regression line
  ggtitle("Retailer Revenue and Store Area with Regression Line") +
  xlab("Store Area 1000 m2") +
  ylab("Revenue million £") +
  scale_colour_manual(name = "Retailer", values = pal) +
  theme_classic()
```

```r
summary(model)
```

Store area explains 89.5% of variation in revenue with p < 0.001. A very strong linear relationship!

Revenue is linear with both store number and area, with the latter having a much stronger relationship, so the better model. Nevertheless there may be information in both variables that might be brought to bear, however, linear regression requires independance between explainatory variables, and store area is expected to correlate with number.

```r
ggplot(data_comparison, aes(x = store_2023, y = area_2023, color = retailer_code)) +
  geom_point() +
  ggtitle("Store Number and Area") +
  xlab("Store Number") +
  ylab("Store Area 1000 m2") +
  scale_colour_manual(name = "Retailer", values = pal) +
  theme_classic()
```

* Most retailers are approximately linear
* The Co-op is an outlier with many small stores.

```r
# Calculate Spearman rank correlation (retained original request)
cor.test(data$store_2023, data$area_2023, method = 'spearman')
```

Spearman's rank is marjinally insignificant, but the high correlation of the majority of values still break the assumption of independance inflating model p-values.

We can run a regession with this caveat.

```r
# Linear model
model <- lm(revenue_2023 ~ store_2023 + area_2023, data = data_comparison)
summary(model)
```

The bivariate model is significant and explains 90.0% of variation in revenue, but store number is not significant in the model so does not add value.

The best model is thus, Revenue ~ Store Area.

# Supermarket Emissions Data

Supermarkets are responsible for carbon emissions from their business operations and value chains. Under the Greenhouse Gas Protocol, emissions are classified as (Fig. 1):

* Scope 1 - direct emissions from company-owned operations.

* Scope 2 - indirect emissions from purchased power.

* Scope 3 - indirect emissions across the retailers entire value chain.

The relative importance of these and other factors vary by business sector, model, and operation.

# UK Emissions Reporting

In the UK, businesses that meet certain criteria are legally required to report a 'streamlined' set of carbon metrics in their annual reports, but may also voluntatrily publish less aggregated data.

The legal framework for emissions and energy reporting began on 1 October 2013 when the Companies Act 2006 [Strategic Report and Directors’ Report) Regulations 2013, SI 2013/1970](https://www.legislation.gov.uk/uksi/2013/1970/contents/made) required all UK-based companies listed on the main market of the London Stock Exchange to report annual Scope 1 and Scope 2 greenhouse gas emissions and energy consumption in their annual reports.

 On 1 April 2019 this was superceeded by the [Streamlined Energy and Carbon Reporting (SECR) regulations (2019)](https://www.gov.uk/government/publications/environmental-reporting-guidelines-including-mandatory-greenhouse-gas-emissions-reporting-guidance) that extended reporting to a wider range of public and private companies, requiring them to report annual:

1. Scope 1 direct emissions as tCO2e.
2. Scope 2 indirect emissions from purchased energy as tCO2e, calculated using either a:
   * *location-based* approach that calculates emissions using the mean carbon intensity of the supplying electrity grid or other system.
   * *market-based* approach that calculates emissions using the carbon intensity of the purchased energy, so discount purchased renewable energy.
3. Total energy consumption as kWh.
4. At least one intensity ratio.

<br>

In addition SECR requires,

* The 'recalculation' of values when a company undergoes major structural change, such as a merger to maintain consitency in reported values through time. Recalculation either sets a new baseline or values may be back-calculated to the previous baseline or other year. In practice, this means several estimates of the same KPI for the same year may be published in the same or different annual reports.

* Transparency is required in methods, comparison to baselines, and the setting of targets.

<br>

Scope 3 emissions reporting is encouraged, but voluntary.

# Supermarket Emissions

Global food systems are responsible for 18 Gigatonnes of CO<sub>2</sub>e equating to 1/3 of global emissions (Crippa et al. 2021) of which the 'retail' component is just 4% (Fig. 2). Within the grocery business itself,
just 7% of emissions arise from direct Scope 1 and 2 operations, with the remaining 93% falling within Scope 3 (Jessie et al. 2022), including upstream food production, processing, and transport, and downstream customer transport and waste. The focus here is on Scope 1 and 2 emissions and energy use as Scope 3 reporting is volunatry and partial.

# Retailer Size

Larger busineses are expected to use more energy and produce more emissions, simply because of the greater size of their operation. Size, however, may also lead to economies of scale increasing efficiency.

SECR require companies to publish at least one intensity metric scaled to a measure of business size, which for supermarkets is most often revenue or a measure of the area of the retailer physical estate. Intensity metrics calculated with different demonimators are not comparable.

## Estate Size
Retailer size may be measured by the physical area covered by business operations. 'Business area' may, however, be calculated as 'sales floor', area of stores or area of stores and distribution centers.



# Store Size

Emissions vary with store size, with larger and discount stores having lower carbon intensities than smaller convenisnece stores (Ferreria et al . 2018).

* Store size impacts Scope 1 and 2 emissions.

# Estate Management and Operation
Emissions and energy consumption depend on how stores are operated (Marrucci et al. 2020; Clarke 2024),  maintained and retrofit (Jenkins 2008), and ulltimately constructed and decommissioned (Blay-Armah et al. 2023; Mohebbi et al. 2023). Electricity may account for upto 60% of direct supermarket emissions (Joint Research Centre 2013).

* Store operation generates Scope 1 and 2 emissions.
* Construction, decomissioning, and retrofit are Scope 3, unless compleated in-house.
* Maintenance may contribute to all Scopes.

# Fugitive Emissions
Fugitive emissions of refridgerants may account for up to 40% of direct supermarket emissions (Joint Research Centre 2013), supported by fugutive emissions accounting for 25% of Asda and 40% of Lidls scope 1 emissions in 2023.

* Fugitive emssions are Scope 1.

# Electricity

Electricity may account for upto 60% of direct supermarket emissions (Joint Research Centre 2013).

It is used to power stores, distirbution centres, and offcies, and increasingly vehicules.

* Purchased electricity is Scope 2.

# Gas

Gas is primarily used to power on-site boilers. In 2023, Asda report gas accounted for 27% of Scope 1 emissions, while Lidl report 27%.

* Gas combusted in boilers is Scope 1

# Transport
How produce is transported from distribution centres to stores and homes influence emissions (Zhang et al. 2023) dependant on the spatial distribution of stores, depots, and customers, and the efficiency of the transport network, vehicles, and routing.

In 2023, transport accounted for 49% of Asda's Scope 1 emissions, 60% of Morrison's, and a whopping 92% of Lidl's.

Online purchases increased 3.7% to 10% of revenue during COVID, only dropping by 0.5-1% in subsequent years (Fig. 1). The proportion of UK supermarket online shopping that is collected versus delivered is not known; however, a 2022 US study reported 63% of customers preferred grocery delivery over click and collect (McKinsey & Company, 2022). Emissions may be expected to grow with increased retailer delivery, with emissions allocated to scope depending on mode of power and ownership.

* Retailer-owned carbon-fulled transport is Scope 1.
* Retailer-owned electic-powered vehicles are Scope 2.
* Customer and contractor emissions are Scope 3 (Category 4 or 9, and Category 11 respectively).



# Products
Supermarkets vary in the variety and cost of products stocked which may impact emissions and energy consumption.

* Scope 1 and 2 emissions arise from storing and transporting products

* Production, use, and disposal are Scope 3.

# System Model

The UK supermarket emissions system is conceptualised factors that generate emissions (drivers), source categories within Scopes, and reported metrics (Fig. 7). 'Scope 2 location residual' is difference between market- and location-based emissions, and are omissions 'offset' through the purchase of renewable zero-carbon energy.

<BR>
<img src="https://github.com/davidmkidd/UK-Supermarket-Carbon-Emissions/blob/main/image/UKSmktComp_Conceptual.png?raw=1" width = "800">
<figcaption>Figure 7. Drivers, sources, scopes, and metrics.</figcaption>
<BR>

Sources, Scopes, and metrics are mathematically related.


* Emissions are the product of an amount and a carbon intensity value.

>> $C = A * I$

>> where *C* is carbon, *A* is amount of a GHG source (e.g. Energy *E*) , and *I* carbon intensity.


* Emission drivers exist across the entire value chain with many expected to increase with 'retailer size'.

* Total value-chain emissions are the sum of Scope 1, Scope 2, and Scope 3 emissions.

>> $C_{total} = C_{scope1} + C_{scope2} + C_{scope3}$

* Most Scope 1 emissions are from three sources that are expected to scale with retailer size,

>> $C_{scope1} \approx  C_{gas} + C_{transport} + C_{fugitive}$

* Most Scope 2 emissions are from electricity purchased from the UK National Grid, so

>> $C_{scope2, location} \approx C_{electricity} * I_{grid}$

>> and,

>> $C_{scope2, market} = E_{electricity} * I_{supplier}$

> Scope 2 location-based emissions are expected to scale with retailer size.

> Scope 2 market-based emissions are **not** expected to scale with retailer size.

* The residual between Scope 2 location and market emissions is carbon generated by operations but 'offset' through the market.

>> $C_{scope2, residual} = C_{scope2, location} - C_{scope2, market}$

>> And,

>> $C_{scope12, residual} = C_{scope12, location} - C_{scope12, market}$

> Location-based scope 2 emissions are greater than market-based values, unless purchased enegy has a higher carbon intensity than the UK National Grid which is unlikely, but not impossible.

* Combined Scope 1 and 2 emissions are emissions generated from a retailer's direct business operations.

>> $C_{scope12, location} = C_{scope1} + C_{scope2, location}$

>> $C_{scope12, market} = C_{scope1} + C_{scope2, market}$

> Scope 2 location-based emissions are expected to scale with retailer size.

> Scope 2 market-based emissions are may scale with retailer size, but is a composite of a scaling and unscaling values.

* Scope 3 emissions are by far the largest component, but reporting voluntary and partial.

* Total Energy is the sum of Scope 1 gas and transport, and scope 2 location-based emissions.

>> $E_{total} = E_{electricity} + E_{gas} + E_{transport}$
